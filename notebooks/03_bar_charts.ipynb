```python
import pandas as pd
import altair as alt
from IPython.display import HTML
```

```python
uri = 'https://vega.github.io/vega-datasets/data/penguins.json'

data = pd.read_json(uri)
data['i'] = data.index
```

```python
data.head()
```

##### Basic bar chart

```python
chart = alt.Chart(data).mark_bar().encode(
    x='count():Q',
    y='Species:N'
)

HTML(chart.to_html())
```

##### Add a title

```python
chart.properties(title='Penguins')
```

##### Add text labels

```python
chart + chart.mark_text(dx=3, align='left').encode(text='count()').properties(title='Penguins')
```

##### Stacked bars

```python
chart.encode(color='Island')
```

##### Add tooltip

```python
chart.mark_bar(tooltip=True).encode(color='Island')
```
