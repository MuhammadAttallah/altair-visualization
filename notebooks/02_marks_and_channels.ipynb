```python
import pandas as pd
import altair as alt
from altair import datum
from vega_datasets import data
```

```python
uri = 'https://vega.github.io/vega-datasets/data/penguins.json'
```

```python
data = pd.read_json(uri)
data['i'] = data.index
```

```python
data.head()
```

#### Magnitude Channels
##### Position Common Scale

```python
alt.Chart(data).mark_circle().encode(
    x = alt.X('Beak Length (mm):Q', scale=alt.Scale(zero=False))
)
```

##### Position on Unaligned Scale

```python
alt.Chart(data).mark_circle().encode(
    x = alt.X('Beak Length (mm):Q', scale=alt.Scale(zero=False)),
    column='Species:N'
)
```

```python
print(alt.Chart(uri).mark_circle().encode(
    x = alt.X('Beak Length (mm):Q', scale=alt.Scale(zero=False)),
    column='Species:N'
).to_json())
```

##### Length

```python
alt.Chart(data).mark_bar().encode(
    x = alt.X('Beak Length (mm):Q', scale=alt.Scale(zero=True)),
    y = 'i:O'
).transform_filter(datum.i < 10)
```

##### Area

```python
alt.Chart(data).mark_circle().encode(
    x = alt.X('Beak Length (mm):Q', scale=alt.Scale(zero=False)),
    size = 'Beak Length (mm):Q'
).properties(height=200, width=700)
```

##### Color Luminance

```python
alt.Chart(data).mark_circle().encode(
    x = alt.X('Beak Length (mm):Q', scale=alt.Scale(zero=False)),
    color = alt.Color('Beak Length (mm):Q', scale=alt.Scale(scheme='greys')),
    size = 'Beak Length (mm):Q'
).properties(height=200, width=700)
```

#### Identity Channels

##### Spatial Region

```python
alt.Chart(data).mark_circle().encode(
    x = alt.X('Species:N'),
    y = alt.Y('Beak Length (mm):Q', scale=alt.Scale(zero=False)),
    color = alt.Color('Beak Length (mm):Q', scale=alt.Scale(scheme='oranges')),
    size = 'Beak Length (mm):Q'
).properties(height=200, width=700)
```

##### Colr Hue

```python
alt.Chart(data).mark_circle().encode(
    x = alt.X('Species:N'),
    y = alt.Y('Beak Length (mm):Q', scale=alt.Scale(zero=False)),
    color = alt.Color('Species:N', scale=alt.Scale(scheme='category10')),
    size = 'Beak Length (mm):Q'
).properties(height=200, width=700)
```
